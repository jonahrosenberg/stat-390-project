# Data Combination and Splitting

This notebook reads `train.csv` and `test.csv` from the `data/raw` folder, combines them into a single dataset, and then splits them back into a new training and testing set in the `data` folder.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
import random

# Define file paths relative to the scripts folder
raw_train_path = '../data/raw/train.csv'
raw_test_path = '../data/raw/test.csv'
output_train_path = '../data/train.csv'
output_test_path = '../data/test.csv'

# 1. Load the raw datasets
train_df = pd.read_csv(raw_train_path)
test_df = pd.read_csv(raw_test_path)

# 2. Combine them into one dataframe
combined_df = pd.concat([train_df, test_df], ignore_index=True)
print(f"Combined dataset has {len(combined_df)} rows and {len(combined_df.columns)} columns.")

# 3. Resplit the combined data into training and testing sets (e.g., 80% train, 20% test)
new_train_df, new_test_df = train_test_split(combined_df, test_size=0.2, random_state=5255)

# Ensure the output directory exists
os.makedirs('../data', exist_ok=True)

# 4. Save the new splits to the general data folder
new_train_df.to_csv(output_train_path, index=False)
new_test_df.to_csv(output_test_path, index=False)

print(f"Saved {len(new_train_df)} rows to {output_train_path}")
print(f"Saved {len(new_test_df)} rows to {output_test_path}")
```
